# Mini Prompt Engine using LangChain

## Objective
Build a modular and reusable prompt system using LangChain instead of hardcoded prompts.

## Key Concepts Covered
- PromptTemplate
- ChatPromptTemplate
- Input Validation
- Prompt Variations
- Modular Design
- Reusability

## Pipeline
User Input → Validation → Template → Prompt Generation → Output

## Imports

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

## Template Registry

We centralize all PromptTemplates in one place to ensure:
- Reusability
- Clean architecture
- Easy scaling

```python
def get_templates():

    return {
        "basic": PromptTemplate(
            input_variables=["topic"],
            template="Explain {topic} in simple terms for beginners"
        ),

        "multi_input": PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} tone"
        ),

        "teaching": PromptTemplate(
            input_variables=["topic"],
            template="Explain {topic} clearly step by step"
        ),

        "interview": PromptTemplate(
            input_variables=["topic"],
            template="Ask 3 questions about {topic}"
        ),

        "storytelling": PromptTemplate(
            input_variables=["topic"],
            template="Explain {topic} as a story"
        )
    }
```

## Replace Hardcoded Prompt

Converted static function into a reusable PromptTemplate.

### Before:
Hardcoded string using f-string

### After:
Dynamic PromptTemplate

```python
templates = get_templates()

print(templates["basic"].format(topic="Artificial Intelligence"))
```

##  Multi-Input Prompt System

We created a PromptTemplate with multiple inputs:
- topic
- audience
- tone

This allows dynamic prompt generation.

```python
print(templates["multi_input"].format(
    topic="AI", audience="beginners", tone="friendly"
))

print(templates["multi_input"].format(
    topic="Python", audience="kids", tone="fun"
))

print(templates["multi_input"].format(
    topic="Deep Learning", audience="engineers", tone="technical"
))
```

## Prompt Variations Engine

We created multiple templates for different use cases:
- Teaching → step-by-step explanation
- Interview → question generation
- Storytelling → narrative explanation

This demonstrates prompt flexibility.

```python
def run_prompt_variations(topic):
    templates = get_templates()

    print("Teaching:\n", templates["teaching"].format(topic=topic))
    print("\nInterview:\n", templates["interview"].format(topic=topic))
    print("\nStorytelling:\n", templates["storytelling"].format(topic=topic))

run_prompt_variations("Machine Learning")
```

## ChatPromptTemplate System

We use role-based prompting:
- System → defines behavior
- User → provides input

Roles implemented:
- Teacher
- Interviewer
- Motivator

```python
def get_chat_prompt(role, topic):

    role_map = {
        "teacher": "You are a helpful teacher who explains clearly.",
        "interviewer": "You are an interviewer asking questions.",
        "motivator": "You motivate and inspire learning."
    }

    system_message = role_map.get(role, "You are a helpful assistant.")

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Explain {topic}")
    ])

    return chat_template.format_messages(topic=topic)

messages = get_chat_prompt("teacher", "Neural Networks")

for msg in messages:
    print(msg)
```

## Input Validation Layer

We validate:
- audience → beginner, intermediate, expert
- tone → formal, casual, fun

Invalid inputs are handled with default values.

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience → default: beginner")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone → default: formal")
        tone = "formal"

    return audience, tone
```

## Prompt Generator Engine

Pipeline:
User Input → Validation → Template → Output

This function integrates the entire system.

```python
def generate_prompt(topic, audience, tone, style):

    templates = get_templates()

    audience, tone = validate_inputs(audience, tone)

    if style not in templates:
        style = "teaching"

    if style == "multi_input":
        return templates["multi_input"].format(
            topic=topic,
            audience=audience,
            tone=tone
        )

    return templates[style].format(topic=topic)

print(generate_prompt(
    "Neural Networks", "beginner", "fun", "storytelling"
))
```

## Template Reusability Test

We use ONE template with multiple inputs.

Goal:
- Same structure
- Different outputs

```python
inputs = [
    {"topic": "AI", "audience": "beginner", "tone": "formal"},
    {"topic": "Python", "audience": "intermediate", "tone": "casual"},
    {"topic": "ML", "audience": "expert", "tone": "fun"},
    {"topic": "DL", "audience": "invalid", "tone": "formal"},
    {"topic": "NLP", "audience": "beginner", "tone": "invalid"},
]

templates = get_templates()

for inp in inputs:
    print(templates["multi_input"].format(
        topic=inp["topic"],
        audience=inp["audience"],
        tone=inp["tone"]
    ))
```
